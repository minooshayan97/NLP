```python
import numpy as np
import pandas as pd
import time
```

```python
f = open("Train.txt", "r", encoding = 'utf-8')
lines = f.readlines()

train_tagged = []

for line in lines:
    if line == '\n':
        continue
    train_tagged.append(tuple(line.strip().split()))
```

```python
tags = {tag for word,tag in train_tagged}

print("Total Number Of Tages In Train Data : {}".format(len(tags)))
print("Tages In Train Data : ")
for tag in tags:
    print(tag) 
```

```python
def word_given_tag(word, tag, train_bag):
    
    tag_list = [pair for pair in train_bag if pair[1]==tag]
    count_tag = len(tag_list)
    w_given_tag_list = [pair[0] for pair in tag_list if pair[0]==word]

    count_w_given_tag = len(w_given_tag_list)
    
    return (count_w_given_tag, count_tag)
```

```python
def p_t2_given_t1(t2, t1, train_bag):
    
    tags = [pair[1] for pair in train_bag]
    count_t1 = len([t for t in tags if t==t1])
    count_t2_t1 = 0
    
    for index in range(len(tags)-1):
        if tags[index]==t1 and tags[index+1] == t2:
            count_t2_t1 += 1
            
    return (count_t2_t1, count_t1)
```

```python
tags_matrix = np.zeros((len(tags), len(tags)), dtype='float64')
start = time.time()
for i, t1 in enumerate(list(tags)):
    for j, t2 in enumerate(list(tags)): 
        p_t2_t1 = p_t2_given_t1(t2, t1, train_tagged)
        tags_matrix[i, j] = p_t2_t1[0]/p_t2_t1[1]
end = time.time()       
tags_df = pd.DataFrame(tags_matrix, columns = list(tags), index=list(tags))
display(tags_df)
end-start
```

```python
f = open("Test.txt", "r", encoding = 'utf-8')
lines = f.readlines()

test_untagged = []
test_tagged = []

for line in lines:
    if line == '\n':
        continue
    tup = tuple(line.strip().split())
    test_untagged.append(tup[0])
    test_tagged.append(tup)
```
